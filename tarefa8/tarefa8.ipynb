```python
import skvideo.io
import numpy as np
import matplotlib.pyplot as plt
import cv2
import math

%matplotlib inline
```

```python
src_video = cv2.VideoCapture('video.mp4', 0)

print('Total frame number', src_video.get(cv2.CAP_PROP_FRAME_COUNT))

video = []
while (src_video.isOpened()):
    ret, frame = src_video.read()
    if(ret == True):
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        video.append(frame)
    else:
        break
src_video.release()
cv2.destroyAllWindows()

# video = np.array(video)
```

```python
def valid_p(i, j, height, width):
    return (i >= 0 and i < height) and (j >= 0 and i < width)

faked = []
print(video[0].shape)
out = skvideo.io.FFmpegWriter("outputvideo.mp4")

for f in range(150):
    print('current = ', f)
    current = video[f]
    future = video[f+1]
    fake_frame1  = current.copy()
    fake_frame2  = current.copy()
    flow = cv2.calcOpticalFlowFarneback(current,future, None, 0.5, 3, 15, 3, 5, 1.2, 0)
    mag, ang = cv2.cartToPolar(flow[...,0], flow[...,1])
    faked.append(current)
    for i in range(ang.shape[0]):
        for j in range(ang.shape[1]):
            if(mag[i, j] != 0):
                angule = ang[i, j]*180/np.pi/2
                magnitude = abs(int(mag[i, j]))
                if(angule >= 0 and angule <= 45) or (angule > 270 and angule <= 360):
                    if(valid_p(i, j+magnitude, ang.shape[0], ang.shape[1])):
                        fake_frame1[i, j+magnitude//2] = current[i, j]
                        fake_frame2[i, j+magnitude] = current[i, j]
                elif(angule > 45 and angule <= 115):
                    if(valid_p(i+magnitude, j, ang.shape[0], ang.shape[1])):
                        fake_frame1[i+magnitude//2, j] = current[i, j]
                        fake_frame2[i+magnitude, j] = current[i, j]
                elif(angule > 115 and angule <= 180):
                    if(valid_p(i, j-magnitude, ang.shape[0], ang.shape[1])):
                        fake_frame1[i, j-magnitude//2] = current[i, j]
                        fake_frame2[i, j-magnitude] = current[i, j]
                elif(angule > 180 and angule <= 270):
                    if(valid_p(i-magnitude, j, ang.shape[0], ang.shape[1])):
                        fake_frame1[i-magnitude//2, j] = current[i, j]
                        fake_frame2[i-magnitude, j] = current[i, j]
    out.writeFrame(current)
    out.writeFrame(np.uint8(fake_frame1))
    out.writeFrame(np.uint8(fake_frame2))
    
out.close()

                
            
            
# current = video[0]
# future = video[1]
# flow = cv2.calcOpticalFlowFarneback(current,future, None, 0.5, 3, 15, 3, 5, 1.2, 0)
# mag, ang = cv2.cartToPolar(flow[...,0], flow[...,1])
# mag
```
